# Introduction to Networking
## 1. Communication
Communication is the process of sharing information, ideas, or feelings between people (or groups) so that they understand each other.
- Using signal
- Via a medium

## 2. Information
Any type of data that has been organized or processed so it has meaning and can be understood.
- Text
- Image
- Audio
- Video

## 3. Medium
A path or channel through which a message travels from sender to receiver.
- Wired
- Wireless

## 4. Signal
A **form** the message takes so it can move through the medium.
- Analog
- Digital
- Electrical
- Optiacal / Light
- Radio Frequency

## 5. Electrical 
The branch of engineering that deals with devices and systems operating on high-power AC (alternating current) electricity, such as:
- AC
- TV
- Refrigerator
- Washing Machine

## 6. Electronic
 The branch of engineering that deals with devices and systems operating *mainly* on low-power DC (direct current) electricity, often using semiconductors to process signals, such as:
 - Mobile phones
 - Laptops/Tabs
 - Smartphones
 - PC/Computer

## 7. Network
Group of devices connected with each other
- Bluetooth file sharing
- Wifi
- Internet

# OSI Model

## Introduction to OSI Model

- Developed by ISO - International Organization for standardization

- Stands for - Open System Interconnect Model

- Idea was to develop vendor independent hardware and software

- Useful tool for network troubleshooting

- **It just explain how internet works and how data flow in network**

- Has no physical appearance in the network

- Only used for teaching purpose

## OSI Architecture (7-layered architecture)

|Layer|Name|
|-----|----|
|7|Application Layer|
|6|Presentation Layer|
|5|Session Layer|
|4|Transport Layer|
|3|Network Layer|
|2|Data Link Layer|
|1|Physical Later|

> **Note:** These 7 layers are furthur divided into 2 more categories for the sake of simplicity: 

1. **Upper Layers:** close to the user

2. **Lower Layers:** close to the network


### 1. Upper Layer

|Layer|Name|
|-----|----|
|7|Application Layer|
|6|Presentation Layer|
|5|Session Layer|

- These are the layers that generally deals with the applications only
- They have no real involvement with the real data packaging or transportation

### 2. Lower Layer

|Layer|Name|
|-----|----|
|4|Transport Layer|
|3|Network Layer|
|2|Data Link Layer|
|1|Physical Later|

- These layers actually deals with how data is really being transferred through the network.
- Vast majority of network engineering is all about these lower layers.
- You would be focusing on upper layers if you are more of a programmer and writing applications. 

> **Note:** Upper layers focus on **programming** and application-level functions, while the Lower layers handle the **actual transfer of data**.  In network engineering, the primary focus is on the lower layers.

## Functions of each OSI  layer

| Layer|Function|
| -----|--------|
| **Application Layer**  | Provides a platform for network access to applications (e.g., web browsers, email clients). Provides network services like HTTP, SMTP, FTB, DNS              |
| **Presentation Layer** | Handles data format, translation, compression and encryption/decryption (e.g., HTTPS encryption).                    |
| **Session Layer**      | Establishes, manages, and terminates sessions between client and server, keeping them isolated. Handles synchronization.          |
| **Transport Layer**    | Provides end-to-end communication with segmentation, flow control, error control, and port addressing (e.g., TCP, UDP), making sure data is sent and received correctly.       |
| **Network Layer**      | Handles logical (IP) addressing, routing, packet sforwarding and selects the best path to the destination.                         |
| **Data Link Layer**    |Facilitates node-to-node data transfer within the same network; handles framing, MAC addressing, flow control, and error detection (includes LLC and MAC sublayers)|
| **Physical Layer**     | Facilitates physical transmission of raw bits; defines hardware, cabling, topology, signal encoding, bit rate, and synchronization|


## 7. Application Layer

- This layer provides an interface between the user and applications to access the network servies.

- Eg.  web browsers, email clients, messaging apps (e.g., WhatsApp).

- This is the layer which Communicates directly with end-user devices and applications.

- Common protocols/services: HTTP, SMTP, FTP, Telnet.

> **Note:** Networking is all about enabling applications to work and communicate over the network.

## 6. Presentation Layer

- This layer makes sure that 2 end devices on the network can talk to each other and understand each other regardless of the operating systems , data formats or encryption types they are using. 

**Other functions**

- Data encryption and decryption

- Data compression and decompression

- Data translation between formats

- Data formatting and syntax handling

## 5. Session Layer

- It handles the session between end hosts and web servers

- Responsible for starting, maintaining, and terminating communication sessions between clients and servers.

- Many web applications rely on sessions to track user activity and maintain connections.

- Eg. When you are logged into Gmail or Facebook, you have an active session with their servers. Logging out ends the session, and logging in again creates a new session.

> **Note:** In Windows CMD, typing ``NETSTAT`` will display the active sessions and connections on your system.

## 4. Transport Layer

- It breaks large chunks of data into smaller units for transmission, since complete data cannot be sent as a single block over the network.

- At the sending side, data is segmented; at the receiving side, these segments are reassembled into the original data.

- This process is called **segmentation**, and the smaller units at this layer are known as **segments**. at transport layer. 

- **It establishes end-to-end connectivity between devices using port numbers.**

- It ensures reliable or unreliable delivery depending on the protocol used (e.g., TCP for reliable, UDP for faster but unreliable).

> **Header:** Each layer adds its own information with the main data in the form of header. It is known as header becuase it is processed before the main data.

- Ensure reliable data delivery via error detection and re-transmission (using TCP)

- Multiplexing and demultiplexing using port numbers.

- **Protocols Used:** TCP, UDP

### TCP
- TCP is used for reliable connectivity (by TCP 3-way handshaking process)
- eg. SMTP, FTP

### UDP
- UDP is used for faster but unreliable communications
- eg. Phone calls, Video Calls

### Port Numbers: Range (0 to 2¹⁶-1) (0 to 65535)
- 0 to 1023 - Well-known ports (assigned to standard services).
- 1024 to 64511 - Registered ports (used for general client-server sessions).
- 64512 to 65535 - Dynamic or private ports.

### Socket = IP Address + Port Numbers
- A unique combination of IP address + port number used to identify a specific connection.

## 3. Network Layer

- Provides connectivity between devices across different networks and determines - the best path for data transmission.

#### Logical Addressing:
- Assigns and uses IP addresses for device identification.
- Supports both IPv4 and IPv6.

#### Devices and Protocol at this layer
- Devices: Router
- Protocols: IPv4, IPv6, ICMP (Internet Control Message Protocol)

> **Note:** In Windows CMD, type ``ipconfig`` to view your device’s IP address details.

## 2. Data Link Layer
- It performs the final framing and formatting of data before sending it over the physical medium.
- Responsible for node-to-node delivery within the same network segment.

#### Define Physical Addressing:
- Uses MAC (Media Access Control) addresses to identify devices on the local network.

#### Control **Error Detection**:
- Uses methods such as **Cyclic Redundancy Check (CRC)** to detect transmission errors.

#### Device and Protocols at this layer:
- **Devices:** Switches, bridges, wireless access points
- **Protocols:** Ethernet, PPP (Point-to-Point Protocol)

## 1. Physical Layer
Handles the transmission of raw bits over a physical medium.
It defines physical properties of the network medium: 
- Electrical/optical signaling methods
- Data transmission rates (bit rates)
- Physical connectors and interfaces
- Maximum cable lengths/distances
- Optical wavelengths
- Wireless frequencies

#### Devices and media:
- Hubs, repeaters
- Ethernet (CAT) cables, fiber optic cables
- Wireless antennas and transmitters

## Why use layered architecture?

#### 1. Devices only need to be aware of their own layer:

- Web servers don’t care whether requests arrive via wired cables or wireless signals.

- Switches don’t care whether they are forwarding IPv4 or IPv6 packets, as that’s handled by a higher layer.

#### 2. Enables interoperability between devices and vendors:

- Google Chrome can communicate with an Apache server because both follow HTML standards.

- A Huawei Ethernet switch can connect to a D-Link Ethernet switch because they follow Ethernet standards.

- A Cisco router can connect to a Juniper router because they follow the same IP routing standards.

## Encapsulation:
- The process of adding headers (and sometimes trailers) at each layer on the sending host to create a PDU (Protocol Data Unit).
- It occurs as data moves down the OSI stack (Layer 7 → Layer 6 → Layer 5 → … → Layer 1).
- At each step, the current layer adds its own control information before passing the data to the layer below.
- The process repeats until the physical layer is reached for transmission.

## Decapsulation:
- The process of removing headers/trailers at each layer on the receiving host to retrieve the original data from the PDU.
- Occurs as data moves up the OSI stack (Layer 1 → Layer 2 → Layer 3 → … → Layer 7).
- At each step, the current layer removes its own control information before passing the data to the layer above.
- Repeats until the Application Layer is reached, delivering the original message to the user.

## Where is OSI helpful in real life?
When troubleshooting a network problem we often go up the OSI model:

**Physical**
- Is the network cable plugged in?
- Is the device powered on?

**Data link**
- Is the link light (LED on the port) active?

**Network**
- Are you getting a valid IP address?

**Transport**
- Can you ping your default gateway?

**Session**
- Do you have valid DNS server information?
- Can you ping an external IP (e.g., 8.8.8.8) but not a domain name (e.g., www.google.com)?

**Presentation & Application**

# TCP / IP Model

- Full form: Transmission Control Protocol / Internet Protocol model

- Purpose: A practical networking model with 4 layers (Link → Internet → Transport → Application).

- Function: Defines how the internet actually works; protocols like TCP, IP, HTTP, FTP, DNS are based on it.

- Implemented; created by DARPA for the ARPANET and now used globally.

## TCP vs UDP

| TCP (Transmission Control Protocol)       | UDP (User Datagram Protocol)         |
| ----------------------------------------- | ------------------------------------ |
| Connection-oriented                       | Connectionless                       |
| Uses 3-way handshake before data transfer | No handshake required                |
| Slower (due to reliability checks)        | Faster (less overhead)               |
| More reliable and secure                  | Less reliable, not inherently secure |
| Supports windowing (flow control)         | No windowing (no flow control)       |

> **Window Size:** Size of data that can be transmitted in a single segment. It can be changed in TCP but not in UDP

## MAC address - Media Access Control
- Hardware or Physical address of a device in a network
- Also known as BIA (Burn in Address)
- It is Globally unique: one MAC address is never reused on another device anywhere in the world
- If the device is faulty or corrupted, that MAC address is wasted.
- It is of 48 bits and is written in **hexadecimal** number system (0–9, A–F). 
- It has 2 parts:  OUI (24 bits) + NIC Specific (24 bits)

| **OUI** (Organizationally Unique Identifier)        | **NIC Specific**                                          |
| --------------------------------------------------- | --------------------------------------------------------- |
| 24 bits — identifies manufacturer (e.g., Dell, IBM) | 24 bits — unique ID for the device from that manufacturer |



**For Example:** If Dell makes 1,000 laptops:
- OUI identifies Dell
- NIC part uniquely identifies each laptop’s MAC address

> **Note:** On windows CMD, type ``getmac`` to get MAC address of the device.

# IP(Internet Protocol)
- Provides logical addressing (IPv4 & IPv6) to uniquely identify devices on a network.
- Responsible for routing packets from source to destination across networks.
- Connectionless: Sends packets (called datagrams) independently, without guaranteeing delivery.

#### **IPv4** 
- It is a 32 bit logical address, represented in **Dotted Decimal Notation**. 
- 4 octets (8-bit blocks) separated by dots (.) e.g., 192.168.1.1
- 32-bit addressing → ~4.3 billion unique addresses
- Address shortage problem

#### **IPv6** 
- It is a 128 bit logical address.
- 8  hextets (16-bit blocks) separated by colons (:) e.g.,2001:0db8:85a3::8a2e:0370:7334
- 128-bit addressing → ~3.4 × 10³⁸ unique addresses (almost unlimited)
- Solves shortage, faster routing, better security

#### IANA - Internet Assigning Number System
- Responsible for managing (assigning) key internet number resources. 
- For ex. IP, MAC, ASN

#### IETF - Internet Engineering Task Force

#### ISOC - Internet Society
- Supports and promotes the development and open use of the internet.

#### IAB - Internet Architecture Board

#### IEEE- Institute of Electrical and Electronic Engineering
- Develops global standards for electrical, electronics, and related technologies.
- Examples: Ethernet (IEEE 802.3), Wi-Fi (IEEE 802.11).

> **Note:**

- IANA divides the world into RIRs (Regional Internet Registries), for IP distribution. 
- Each RIR is responsible for distributing IP addresses within its region.
- India falls under **APNIC** – Asia Pacific Network Information Centre.

#### Network ID
- If given, **IP address of a device** is ``10.25.37.89`` (Class A)

- Then its **Network ID** would be ``10.0.0.0`` (all Host bit =0)

- That means any IP from ``10.0.0.1 to 10.255.255.254`` is part of that same network.

#### Broadcast
- When **all host bits are set to 1**, the address becomes the broadcast address.

- If **Network address** is ``10.0.0.0``

- Then ``10.255.255.255`` can be used as the destination IP to broadcast to all hosts within that network.

## What is Subnet Mask?
* A subnet mask is a **32-bit number** that works alongside an IP address.

* It tells us, which part of the IP address represents the **network** and which part represents the **host**.

* In subnet mask, all **networks bits = 1** and all **host bits = 0**

### Reserved IP Addresses

1. **Default Route** : ``0.0.0.0``
Used to represent “any” address when a device doesn’t yet have an IP or when setting up routing rules.

2. **Loopback Address** : ``127.0.0.1``
Used by a computer to talk to itself for testing and troubleshooting.

3. **Link Local Routing** : ``169.254.0.0``
Automatically assigned when a device can’t get an IP from a DHCP server. Works only within the local link.

4. **Multicast Address** : ``224.0.0.0 to 239.255.255.255``
Used to send data to multiple devices at once.

5. **Future Usable Addess** : ``240.0.0.0 to 255.255.255.255``
Set aside for potential future networking purposes.

6. **Universal IP Address** : ``127.0.0.0``

### Loopback IP Address

- The loopback IP (127.0.0.1) is reserved for local host communication.
- Loopback helps in troubleshooting - It’s a way to check if your computer’s TCP/IP stack — the networking brain of your OS — is working properly.
- 127.0.0.1 works without any network connection because it’s purely internal to your computer.

#### How to check if my local host is connection ready (Loopback Testing)?
Open Command Prompt on Windows. Type ``ping 127.0.0.1``

- If all packets are sent and received successfully → your network software is working.

- If packets fail → something’s wrong with your local TCP/IP configuration.


#### What if loopback testing fails?

- If the loopback test fails, it means your computer’s TCP/IP stack is broken or disabled.

- Without a working TCP/IP stack, your system can’t use IP networking at all — so you won’t be able to assign a functional IP address, connect to other devices, or access the internet.

- This is true **even** if your physical network connection (Wi-Fi/Ethernet) is fine.

#### What does `ping` do? 
Think of ``ping`` like sending a short “Hello, are you there?” message to an IP address.

- If the address replies → it’s reachable.

- If there’s no reply → the device is offline, unreachable, or the address is invalid.
